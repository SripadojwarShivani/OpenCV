```python
import cv2 as cv
import numpy as np
```

```python
image=cv.imread('flower.png')
```

```python
height,width=image.shape[:2]
```

```python
#applying matrix to the image
matrix=cv.getRotationMatrix2D((width/2,height/2),10,1)
```

```python
trans=cv.warpAffine(image,matrix,(width,height))
```

```python
cv.imshow('trans',trans)
cv.waitKey(4000)
```
