```python
import cv2 as cv
import numpy as np
```

```python
RED = (0, 0, 255)
p0 = (10, 100)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
img = np.zeros((200, 500, 3), np.uint8)
cv.putText(img,'Computer Vision', p0, font, 0.5, RED, 1, cv.LINE_AA)
cv.imshow('window', img)
cv.waitKey(0)
cv.destroyAllWindows()
```
