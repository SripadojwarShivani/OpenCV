```python
import cv2
import numpy as np
```

```python
image=cv2.imread("flower.png")
```

```python
#applying translation matrix
matrix=np.float32([[1,0,100],[0,1,100]])
```

```python
#applying matrix to the image
trans=cv2.warpAffine(image,matrix,(image.shape[1]+100,image.shape[0]+100))
```

```python
cv2.imshow('translated',trans)
cv2.waitKey(4000)
```
