```python
import cv2
import numpy as np
```

```python
#reading the image
image = cv2.imread('/home/shivanisri/Desktop/1.jpg')
#sharpening image using array
sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
#apply filter
sharpen = cv2.filter2D(image, -1, sharpen_kernel)
```

```python
#display the image
cv2.imshow('sharpen', sharpen)
cv2.waitKey()
```
