```python
import cv2
import numpy as np
```

```python
img=cv2.imread("flower.png")
```

```python
resize=cv2.resize(img,(30,30))
```

```python
img_linear=cv2.resize(img,None,fx=2.5,fy=2.5,interpolation=cv2.INTER_LINEAR) #fx and fy are height and width, low contrast
```

```python
img_cubic=cv2.resize(img,None,fx=2.5,fy=2.5,interpolation=cv2.INTER_CUBIC) #contrast is more
```

```python
cv2.imshow('original',img)
cv2.imshow('linear',img_linear)
cv2.imshow('cubic',img_cubic)
```

```python
if(cv2.waitKey()==ord('q')):
    cv2.destroyAllWindows()
```
